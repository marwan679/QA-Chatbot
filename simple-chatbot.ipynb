# LLM Integration

```python
llm = HuggingFaceEndpoint(repo_id="mistralai/Mistral-7B-v0.1")
```

# Inference

```python
query:str = "What is the capital of France?"
response = llm.invoke(query,
                      temperature=0.5, 
                      max_new_tokens=50)
print(response)
```
